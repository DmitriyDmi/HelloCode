```python
def num_add(str):
  for i in range(len(str)):
    if str[:len(str)-i].isdigit():
      new_str = str[:len(str)-i]
      return new_str
```

```python
def class_add(str):
  for i in range(len(str)):
    if str[:len(str)-i].isdigit():
      new_str = str[-i:]
      return new_str
```

```python
def link_add(str):
  for i in range(len(str)):
    if str[:len(str)-i].isdigit():
      new_str = 'https://codeforces.com/problemset/problem/' + str[:len(str)-i] + '/' + str[-i:]
      return new_str


link_add('18H1')
```

```python
data['link'] = data['#'].apply(link_add)
data['number'] = data['#'].apply(num_add)
data['class'] = data['#'].apply(class_add)
data = data.drop(['Unnamed: 2',	'Unnamed: 3',	'Unnamed: 4'], axis = 1)
data.head()
```
